```python
import pandas as pd

df = pd.read_csv("../data/HR-Employee-Attrition.csv")

df.head()
```

```python
attrition_count = df["Attrition"].value_counts()

print(attrition_count)
```

```python
attrition_rate = (
    df["Attrition"].value_counts(normalize=True) * 100
)

print(attrition_rate)
```

```python
import matplotlib.pyplot as plt

df["Attrition"].value_counts().plot(
    kind="bar"
)

plt.title("Employee Attrition")
plt.xlabel("Attrition")
plt.ylabel("Number of Employees")

plt.tight_layout()

plt.savefig("../visuals/attrition_distribution.png")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(10,6))

sns.countplot(
    data=df,
    x="Department",
    hue="Attrition",
    palette="Blues"
)

plt.title("Attrition by Department", fontsize=14)
plt.xlabel("Department")
plt.ylabel("Employee Count")

plt.tight_layout()

plt.savefig("../visuals/department_attrition.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="OverTime",
    hue="Attrition",
    palette="Oranges"
)

plt.title("Overtime Impact on Attrition", fontsize=14)
plt.xlabel("Overtime")
plt.ylabel("Employee Count")

plt.tight_layout()

plt.savefig("../visuals/overtime_attrition.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="JobSatisfaction",
    hue="Attrition",
    palette="Purples"
)

plt.title("Job Satisfaction vs Attrition", fontsize=14)
plt.xlabel("Job Satisfaction Level")
plt.ylabel("Employee Count")

plt.tight_layout()

plt.savefig("../visuals/job_satisfaction_attrition.png")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x="MonthlyIncome",
    bins=30,
    kde=True,
    color="#16A085"
)

plt.title("Monthly Income Distribution", fontsize=14)
plt.xlabel("Monthly Income")
plt.ylabel("Number of Employees")

plt.tight_layout()

plt.savefig("../visuals/monthly_income_distribution.png")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x="Age",
    bins=20,
    kde=True,
    color="#6C5CE7"
)

plt.title("Employee Age Distribution", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Number of Employees")

plt.tight_layout()

plt.savefig("../visuals/age_distribution.png")

plt.show()
```

```python
total_employees = len(df)

attrition_rate = (
    df["Attrition"].value_counts(normalize=True)["Yes"] * 100
)

avg_income = df["MonthlyIncome"].mean()

print("Total Employees:", total_employees)
print("Attrition Rate:", round(attrition_rate, 2), "%")
print("Average Monthly Income:", round(avg_income, 2))
```

# HR Analytics Insights

## Employee Attrition

A significant number of employees left the company, indicating potential retention challenges.

## Department Analysis

Attrition rates vary across departments.
Certain departments may require targeted retention strategies.

## Overtime Impact

Employees working overtime appear more likely to leave the organization.

## Job Satisfaction

Lower job satisfaction levels are associated with higher attrition.

## Income Analysis

Monthly income distribution suggests compensation differences across employees.
Lower-income employees may be more likely to leave.

## Age Analysis

Attrition appears more common among specific age groups.
